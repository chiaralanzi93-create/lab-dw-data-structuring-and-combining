```python
def standardize_format (value):
    if '/' in str(value):
        # Already in 'x/x/xx' format, return as is
        return value
    else:
        # Convert single digit/number to 'x/0/00' format
        return f"{value}/0/00"

# Apply the transformation to the column
df3['Number of Open Complaints'] = df3['Number of Open Complaints'].apply(standardize_format)
```

# Challenge 2: Structuring Data

In this challenge, we will continue to work with customer data from an insurance company, but we will use a dataset with more columns, called marketing_customer_analysis.csv, which can be found at the following link:

https://raw.githubusercontent.com/data-bootcamp-v4/data/main/marketing_customer_analysis_clean.csv

This dataset contains information such as customer demographics, policy details, vehicle information, and the customer's response to the last marketing campaign. Our goal is to explore and analyze this data by performing data cleaning, formatting, and structuring.

```python
df_marketing = pd.read_csv('https://raw.githubusercontent.com/data-bootcamp-v4/data/main/marketing_customer_analysis_clean.csv')
df_marketing.columns
```

```python
df_marketing['income']
```

```python

# Create the pivot table based on income 
pivot_table = df_marketing.pivot_table(
    index='sales_channel', 
    values='income', 
    aggfunc='sum'
)


# Round the total revenue to 2 decimal points
pivot_table = pivot_table.round(2)

total_income = pivot_table['income'].sum()
pivot_table['percentage'] = (pivot_table['income'] / total_income) * 100
pivot_table['percentage'] = pivot_table['percentage'].round(2)

# Display the pivot table
print(pivot_table)
```

```python
# Agent major sales channel with 37% of revenues
# Web is underperforming with only 15% of revenue shares
```

```python


# Create the pivot table with a list for multiple index columns
pivot_table2 = df_marketing.pivot_table(
    index=['gender', 'education'],  # Use a list here
    values='customer_lifetime_value', 
    aggfunc='mean'  # Correctly specify the aggregation function for averaging
)

# Round the result to 2 decimal points
pivot_table2 = pivot_table2.round(2)

# Sort the pivot table by 'customer_lifetime_value' in descending order
pivot_table2 = pivot_table2.sort_values(by='customer_lifetime_value', ascending=False)



# Display the pivot table
print(pivot_table2)
```

Show it in a long format table.

*In data analysis, a long format table is a way of structuring data in which each observation or measurement is stored in a separate row of the table. The key characteristic of a long format table is that each column represents a single variable, and each row represents a single observation of that variable.*

*More information about long and wide format tables here: https://www.statology.org/long-vs-wide-data/*

```python
summary_table = df_marketing.groupby(['policy_type', 'month'])['total_claim_amount'].sum().reset_index()
```
